<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">

    در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی که در فصل‌های گذشته آموختید، استفاده کنید.
    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.

```python
boys_names_lower = [
    "amir","ali","reza","hossein","mohammad","javad","mostafa","mahdi","hadi","vahid",
    "farhad","shahin","kourosh","kian","navid","nima","pouya","pejman","saeed","soheil",
    "shervin","behnam","babak","omid","sina","aria","piroz","saman","samad","salar",
    "shayan","kaveh","kamran","khashayar","keyvan","nader","nemat","yadollah","yousef","ehsan",
    "esmail","hooman","farid","parham","behrang","kambiz","armin","arash","keykhosrow","borna",
    "kia","matin","mehrdad","dariush","davood","parsa","pouyan","sahand","peyman","afshin",
    "anoushirvan","atabak","jamshid","rostam","siavash","bijan","bahram","fereydoon","shahrokh","shahram",
    "masoud","mansour","mohsen","mojtaba","naser","rahim","karim","abbas","qasem","mehdi",
    "majid","hamid","jahan","behrouz","omran","shahriar","shahbaz","payam","farzin","keyhan",
    "homayoun","tirdad","shapour","mahan","ramin"
]
girls_names_lower = [
    "fatemeh","zahra","maryam","narges","leila","mahin","shirin","simin","taraneh","azar",
    "ladan","mahsa","sahar","samira","parisa","parvaneh","fariba","forough","roxana","shahrzad",
    "yasmin","yasaman","nazanin","niloofar","negin","homa","elham","bahar","sepideh","mitra",
    "mehrnaz","mahdokht","anahita","arezoo","negar","golnaz","golzar","golshid","golshan","golnoosh",
    "shaghayegh","shokoufeh","nastaran","mahbanoo","tarla","pegah","pouneh","shima","shohreh","marjan",
    "hengameh","mozhgan","donya","setareh","sana","hadis","reyhaneh","mahdieh","hanieh","soraya",
    "nahid","goli","farzaneh","farnaz","nasrin","mona","rana","raha","roksana","sahba",
    "mahroo","bita","tina","laleh","elaheh","shiva","firoozeh","melika","mania","yasna",
    "ziba","atefeh","katayoun","kiana","kimia","aida","afsaneh","atoosa","sanaz","sanam",
    "mandana","pardis","darya","neda","naghmeh","yalda"
]


import re

def classify_name(row):
    combined = str(row['fullname']) + " " + str(row['username'])
    words = re.findall(r'[A-Za-z\u0600-\u06FF]+', combined)
    words_lower = [w.lower() for w in words]
    
    if any(word in boys_names_lower for word in words_lower):
        return 1
    elif any(word in girls_names_lower for word in words_lower):
        return 0
    else:
        return 2


```

```python
# do some preprocessing :)
import emoji
train['emoji'] = train['biography'].apply(emoji.emoji_count)
train['text_len'] = train['biography'].apply(len)
train['name_g'] = train.apply(classify_name,axis=1)
train.drop(['username','fullname','biography','is_private','is_verified'],axis=1,inplace=True)
train['is_business'].fillna(0,inplace=True)

test['emoji'] = test['biography'].apply(emoji.emoji_count)
test['text_len'] = test['biography'].apply(len)
test['name_g'] = test.apply(classify_name,axis=1)
test.drop(['username','fullname','biography','is_private','is_verified'],axis=1,inplace=True)

```

    حال که داده را پاکسازی کرده و احتمالا ویژگی‌هایی را به آن افزوده یا از آن حذف کرده‌اید، وقت آن است که مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند.
</font>

```python
from sklearn.model_selection import StratifiedKFold

X = train.drop(['gender'], axis=1)
Y = train['gender']
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for train_index, val_index in skf.split(X,Y):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = Y.iloc[train_index], Y.iloc[val_index]
```

```python
# modeling :)
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(learning_rate=1,n_estimators=300)
model.fit(X_train,y_train)
y_pred = model.predict(X_val)
```
