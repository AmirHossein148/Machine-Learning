```python
train_data = pd.read_csv('../data/news_train.csv')
train_data_text_vectors = ss.load_npz('../data/news_train_text_vectors.npz')

test_data = pd.read_csv('../data/news_test.csv')
test_data_text_vectors = ss.load_npz('../data/news_test_text_vectors.npz')
```

    در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی که در فصل‌های گذشته آموختید، استفاده کنید.
    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.
</font>
</p>

```python
train_data_text_vectors = pd.DataFrame(train_data_text_vectors.toarray(),\
                                  columns = ['token' + str(i) for i in range(train_data_text_vectors.shape[1])] )

train_data_text_vectors
                                
```

```python
test_data_text_vectors = pd.DataFrame(test_data_text_vectors.toarray(),\
                                  columns = ['token' + str(i) for i in range(test_data_text_vectors.shape[1])] )

test_data_text_vectors          
```

```python
from sklearn.model_selection import train_test_split
train , validation = train_test_split(train_data, random_state=313, \
                                                   stratify=train_data.label)
```

```python
train_text_vectors = train_data_text_vectors.loc[train.index]
valid_text_vectors = train_data_text_vectors.loc[validation.index]
```

```python
class Preprocessor : 
    def __init__ (self, ) : 
             pass 
            
    def detect_suspicious_site_urls(self, thresh = 0.5) :
        site_url_status = self.df.replace({'label': {'Fake':1, 'Real':0}}).groupby('site_url').label.agg('mean')
        suspicious_sites = site_url_status[site_url_status >= thresh].index.tolist()
        not_suspicious_sites = site_url_status[site_url_status < thresh].index.tolist()
        
        return suspicious_sites, not_suspicious_sites
    
    def detect_suspicious_types(self, thresh = 0.5) : 
        type_status = self.df.replace({'label': {'Fake':1, 'Real':0}}).groupby('type').label.agg('mean')
        suspicious_types = type_status[type_status >= thresh].index.tolist()
        not_suspicious_types = type_status[type_status < thresh].index.tolist()
        
        return suspicious_types, not_suspicious_types
    
    def detect_suspicious_authors(self, thresh= 0.5) :  
        author_status = self.df.replace({'label': {'Fake':1, 'Real':0}}).groupby('author').label.agg('mean')
        suspicious_authors = author_status[author_status >= thresh].index.tolist()
        not_suspicious_authors = author_status[author_status < thresh].index.tolist()
        
        return suspicious_authors, not_suspicious_authors
    
    def transform(self, dataframe: pd.DataFrame, is_train=False) :   
        self.df = dataframe.copy()
        
        if is_train : 
            self.suspicious_sites , self.not_suspicious_sites = self.detect_suspicious_site_urls()
            self.suspicious_types, self.not_suspicious_types = self.detect_suspicious_types()
            self.suspicious_authors, self.not_suspicious_authors = self.detect_suspicious_authors()
              
        if not is_train and (not hasattr(self, 'suspicious_sites') or not hasattr(self, 'suspicious_types') \
                            or not hasattr(self, 'suspicious_authors')) : 
            raise AttributeError('suspicious fields required! call `tranform` method with `is_train` param set to True')
        
        self.df['is_site_url_suspicious'] = None 
        self.df.loc[self.df.site_url.isin(self.suspicious_sites), 'is_site_url_suspicious'] = 1 
        self.df.loc[self.df.site_url.isin(self.not_suspicious_sites), 'is_site_url_suspicious'] = -1
        self.df.is_site_url_suspicious.fillna(0, inplace=True)
        
        
        self.df['is_type_suspicious'] = None 
        self.df.loc[self.df.type.isin(self.suspicious_types), 'is_type_suspicious'] = 1 
        self.df.loc[self.df.type.isin(self.not_suspicious_types), 'is_type_suspicious'] = -1
        self.df.is_type_suspicious.fillna(0, inplace=True)
        
        
        self.df['is_author_suspicious'] = None 
        self.df.loc[self.df.author.isin(self.suspicious_authors), 'is_author_suspicious'] = 1 
        self.df.loc[self.df.author.isin(self.not_suspicious_authors), 'is_author_suspicious'] = -1
        self.df.is_author_suspicious.fillna(0, inplace=True)
        
        self.df.drop(['author', 'published', 'site_url', 'type'], axis=1, inplace=True)
        
        return self.df
        
        
        
```

```python
preprocessor = Preprocessor()
```

```python
preproccesed_train = preprocessor.transform(train, is_train=True)
preproccesed_valid = preprocessor.transform(validation,)
```

```python
preproccesed_train
```

```python
dr = PCA(n_components=500)
reduced_train_text_vectors = dr.fit_transform(train_text_vectors)
```

```python
text_model = SVC()

text_model.fit(reduced_train_text_vectors, preproccesed_train.label)
```

```python
y_pred = text_model.predict(dr.transform(valid_text_vectors))
```

```python
from sklearn.metrics import classification_report, f1_score

print(f1_score(preproccesed_valid.label, y_pred, average='weighted'))
print(classification_report(preproccesed_valid.label, y_pred))
```

```python
preproccesed_train['label_from_text_model'] = text_model.predict(reduced_train_text_vectors)
preproccesed_train.label_from_text_model.replace({
    "Real": 0,
    "Fake" : 1
}, inplace=True)
preproccesed_train
```

```python
model = SVC()
model.fit(preproccesed_train.drop('label', axis=1), preproccesed_train.label)
```

```python
preproccesed_valid['label_from_text_model'] = text_model.predict(dr.transform(valid_text_vectors))
preproccesed_valid.label_from_text_model.replace({
    "Real": 0,
    "Fake" : 1
}, inplace=True)
preproccesed_valid
```
