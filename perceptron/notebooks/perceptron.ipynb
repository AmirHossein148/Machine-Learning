<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn">
    به منظور پیاده‌سازی مدل،‌ کلاسی با نام <code>Perceptron</code> تعریف خواهیم کرد که در هنگام ساخته شدن (<code>__init__</code>) یک ویژگی مرتبط با وزن‌ها را با نام <code>weights</code> به این کلاس اضافه می‌کند. بنابراین در هنگام پیاده‌سازی توابع جهت دسترسی به وزن‌های پرسپترون می‌توانید از <code>self.weights</code> استفاده کنید.
    <br>
    اکنون هرکدام از توابع مورد نیاز این کلاس را شرح خواهیم داد تا شما به پیاده‌سازی آن‌ها بپردازید:
    <br>
    <ul dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
        <li>
        تابع <code>weighting</code> مقادیر ویژگی‌های یک نمونه را گرفته و آن را در بردار وزن ضرب نقطه‌ای می‌کند. سپس ورودی وزن‌دار شده به عنوان خروجی برگردانده می‌شود.
        </li>
        <li>
        در تابعی به نام <code>activation</code> تابع فعال‌ساز را که در اینجا تابع علامت است پیاده خواهیم کرد. این تابع، ورودی وزن‌دار را گرفته و در صورتیکه حاصل بیشتر یا مساوی صفر بود خروجی <code>1</code> و در غیر این‌صورت خروجی <code dir=ltr>-1</code> تولید می‌کند.
        </li>
        <li>
        تابع <code>predict</code> آرایه‌ای از نمونه‌ها را گرفته و آرایه‌ای شامل برچسب پیش‌بینی‌شده برای هرکدام از نمونه‌ها را بر می‌گرداند. در این تابع باید به ازای هر نمونه ابتدا آن را به کمک تابع <code>weighting</code> وزن دار کنید و حاصل را به تابع <code>activation</code> بدهید تا برچسب <code dir=ltr>+1</code> یا <code dir=ltr>-1</code> را برای نمونه تولید کند.
        </li>
        <li>
        الگوریتم اصلی پرسپترون در واقع در تابع <code>fit</code> پیاده می‌شود. این تابع ورودی‌ها و هایپرپارامترهای زیر را گرفته و سپس طبق مراحلی که شرح خواهیم داد شروع به آموزش [وزن‌های] پرسپترون می‌کند.
        </li>
    </ul>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium; padding-right:40px;">
<font face="vazirmatn">
در درسنامه‌ی پرسپترون اشاره کردیم که عبارت بایاس را می‌توان به کمک اضافه کردن یک ورودی ثابت <code>1</code> به ویژگی‌های ورودی تعبیه کرد. بنابراین در تابع <code>fit</code> ابتدا به اول ویژگی‌های هرکدام از نمونه‌ها عدد <code>1</code> را اضافه کنید (به عنوان مثال برای نمونه‌ی <code>inputs[i]</code>، باید <code>inputs[i][0]</code> برابر <code>1</code> باشد). با این کار یک واحد به طول بردار ویژگی نمونه‌ها اضافه خواهد شد. (فراموش نکنید که این کار را در هنگام پیش‌بینی یعنی تابع predict نیز باید انجام دهید.)
<br>
پس از آن نیاز است تا بردار وزن یعنی <code>self.weights</code> را ساخته و مقداردهی کنید. برای مقداردهی اولیه می‌توانید از اعداد تصادفی بین <code dir=ltr>0</code> تا <code dir=ltr>1</code> استفاده کنید.
<br>
سپس در یک حلقه به ازای هر دور (به تعداد <code>epochs</code>) باید یک‌بار کل مجموعه‌ی آموزشی را بررسی کنید. بنابراین در داخل این حلقه، یک حلقه‌ی دیگر خواهید داشت که هربار یکی از نمونه‌ها را بررسی می‌کند (به دلیل استفاده از روش کاهش گرادیان تصادفی در هر تکرار تنها یک نمونه بررسی و وزن‌ها آپدیت می‌شوند).
<br>
به ازای هر نمونه، ابتدا به کمک تابع <code>weighting</code> بردار ویژگی آن را در بردار وزن ضرب نقطه‌ای می‌کنیم تا ورودی وزن‌دار به دست آید. سپس حاصل به‌دست‌آمده را به تابع فعال‌سازی می‌دهیم تا مقدار خروجی مدل را تولید کند. حال با تفریق پیش‌بینی مدل از مقدار (برچسب) واقعی می‌توانیم اختلاف‌شان را به دست آوریم. سپس طبق فرمول‌های معرفی‌شده در درسنامه، با ضرب این اختلاف در نرخ یادگیری و ویژگی‌های ورودی و سپس جمع آن با مقادیر قبلی بردار وزن می‌توانیم به آپدیت بردار وزن بپردازیم.
<br>
پیشنهاد می‌شود در آخر هر دور نسبت به چاپ دقت مدل کنونی بر روی مجموعه‌ی آموزشی نیز اقدام کنید. برای این کار کافیست نمونه‌ها را به تابع <code>predict</code> بدهید تا پیش‌بینی مدل تولید شود و سپس به کمک تابع آماده‌ی <code>accuracy_score</code> دقت آن را به دست آورید.
<br>
<b>نکته: </b>
در هیچ‌کدام از توابع خود به غیر از تابع <code>fit</code> چیزی را چاپ (<code>print</code>) نکنید، در غیر این‌صورت سیستم داوری قادر به امتیازدهی به توابع شما نخواهد بود.
</font>
</p>

```python
class Perceptron:
    def __init__(self):
        self.weights = None
        self.bias = 0.0

    def weighting(self, x):
        return np.dot(x, self.weights) + self.bias

    def activation(self, z):
        return np.where(z >= 0, 1, -1)

    def predict(self, inputs):
        X = np.array(inputs, dtype=float)
        predictions = []
        for x in X:
            pred = self.activation(self.weighting(x))
            predictions.append(pred)
        return np.array(predictions)

    def fit(self, inputs, outputs, learning_rate=0.01, epochs=100):
        X = np.array(inputs, dtype=float)
        y = np.array(outputs)
        
        self.weights = np.random.uniform(-0.01, 0.01, size=X.shape[1])
        self.bias = 0.0

        for epoch in range(epochs):
            for i in range(len(X)):
                x_i = X[i]
                pred = self.activation(self.weighting(x_i))
                error = y[i] - pred

                self.weights += learning_rate * error * x_i
                self.bias += learning_rate * error

```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazirmatn; font-size:medium">
    در این حالت می‌خواهیم کلاسی با نام <code>Adaline</code> بسازیم و  مدل را با قاعده‌ی دلتا آموزش دهیم. در اینجا از روش نزول گرادیان استفاده خواهیم کرد، یعنی تمام نمونه‌های آموزشی به‌صورت همزمان بر روی آپدیت وزن تاثیر می‌گذارند (برخلاف روش نزول گرادیان تصادفی که هربار یک نمونه دیده می‌شد). کد این مدل بسیار شبیه به حالت قبلی خواهد بود و تنها نیاز است چند مورد جزئی زیر را اعمال کنید.
    در کلاس <code>Adaline</code> همانند حالت قبل، هنگام ساخته شدن (<code>__init__</code>)، بردار وزن‌ها مقداردهی اولیه می‌شود. 
    <br><br>
    اکنون هرکدام از توابع این کلاس را شرح می‌دهیم:
    <br>
    <ul dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazirmatn; font-size:medium">
        <li>
        <b>تابع <code>weighting</code></b>: ورودی‌ها  را دقیقا مشابه به حالت قبل را در بردار وزن ضرب نقطه‌ای کرده و مقدار وزن‌دار شده را خروجی می‌دهد.
        </li>
        <li>
        <b>تابع <code>activation</code></b>:         از آنجا که در این حالت از تابع فعال‌سازی خطی استفاده می‌شود تابع <code>activation</code> تنها همان ورودی وزن‌دار خود را بدون هیچ‌گونه تغییری خروجی می‌دهد.
        </li>
        <li>
        <b>تابع <code>predict</code></b>: ورودی‌ها را گرفته و مقدار پیش‌بینی‌شده را برمی‌گرداند. در این تابع، ابتدا مقدار ۱ (بایاس) به ابتدای هر نمونه اضافه شده، سپس مقدار ورودی وزن‌دار شده محاسبه و از تابع فعال‌سازی عبور می‌کند. در نهایت، مقدار خروجی بررسی می‌شود: اگر مقدار آن بزرگ‌تر یا مساوی صفر باشد، مقدار پیش‌بینی <code>+1</code> و در غیر این‌صورت <code>-1</code> خواهد بود.
        </li>
        <li>
        <b>تابع <code>fit</code></b>: الگوریتم اصلی <code>Adaline</code> در این تابع پیاده‌سازی می‌شود. فرآیند یادگیری با روش نزول گرادیان انجام شده و وزن‌ها در هر تکرار به‌روزرسانی می‌شوند.
        </li>
    </ul>

|ورودی|توضیحات|

<p dir="rtl" style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium; padding-right:40px;">
<font face="vazirmatn">
    در ابتدای تابع <code>fit</code>، به هر ویژگی ورودی مقدار ۱ اضافه می‌شود تا بایاس (bias) به داده‌ها وارد شود. این کار به مدل کمک می‌کند که بر اساس داده‌های ورودی تنها نتایج را پیش‌بینی نکند، بلکه بایاس هم در نظر گرفته شود. سپس وزن‌ها به‌طور تصادفی در بازه <code>[0,1]</code> مقداردهی می‌شوند تا مدل در ابتدا شروع به یادگیری از روی مقادیر تصادفی کند.
    <br>
    در هر تکرار (<code>epoch</code>)، ابتدا وزن‌ها به ورودی‌های داده‌ها اعمال می‌شوند تا خروجی وزن‌دار (weighted input) به دست آید. سپس خروجی پیش‌بینی شده مدل از مقدار واقعی (یعنی مقدار <code>outputs</code>) تفریق می‌شود. تفاوت بین این دو، که به آن <b>خطا</b> می‌گوییم، به مدل نشان می‌دهد که چقدر پیش‌بینی‌ها از حقیقت دور هستند.
    <br>
    در Adaline، برخلاف مدل پرسپترون که فقط از علامت خطا استفاده می‌کند (یعنی فقط بررسی می‌کند که آیا پیش‌بینی درست بوده یا نه)، این مدل از مقدار دقیق خطا برای تنظیم وزن‌ها استفاده می‌کند. این به این معنی است که اگر مدل خطای بزرگتری داشته باشد، وزن‌ها بیشتر تغییر می‌کنند.
    <br>
    در نهایت، به‌جای اینکه برای هر نمونه به‌طور جداگانه خطا محاسبه کنیم و وزن‌ها را تغییر دهیم، این محاسبات به‌صورت برداری انجام می‌شود. این یعنی اینکه خطاهای تمام نمونه‌ها به‌طور همزمان محاسبه شده و به‌صورت یک ماتریس ضرب شده تا وزن‌ها به‌طور همزمان به‌روزرسانی شوند. به این صورت که:
    <br>
    <ul dir="rtl" style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
        <li>
            ابتدا <code>weighted_input = np.dot(new_inputs, self.weights)</code> محاسبه می‌شود که در آن <code>new_inputs</code> ماتریس ورودی‌ها با بایاس (ستون ۱) و <code>self.weights</code> بردار وزن‌ها است.
        </li>
        <li>
            سپس <code>diff = outputs - weighted_input</code> محاسبه می‌شود که خطای هر نمونه به‌صورت تفاضل مقدار واقعی (برچسب‌های <code>outputs</code>) و مقدار پیش‌بینی‌شده به دست آمده از ضرب ورودی‌ها در وزن‌ها به‌دست می‌آید.
        </li>
        <li>
            در مرحله‌ی بعدی برای به‌روزرسانی وزن‌ها، از ضرب نقطه‌ای <code>new_inputs.T.dot(diff)</code> استفاده می‌شود. این عمل نشان می‌دهد که هر خطای هر نمونه با ورودی‌های آن نمونه ضرب می‌شود تا تغییرات وزن‌ها بر اساس این خطاها اعمال شود.
        </li>
    </ul>
</font>
</p>



```python
class Adaline:

    def __init__(self):
        self.weights = None
        self.bias = 0.0

    def weighting(self, x):
        return np.dot(x, self.weights) + self.bias

    def activation(self, z):
        return z

    def predict(self, inputs):
        X = np.array(inputs, dtype=float)
        predictions = []
        for x in X:
            pred = self.activation(self.weighting(x))  
            predictions.append(1 if pred >= 0 else -1)
        return np.array(predictions)


    def fit(self, inputs, outputs, learning_rate=0.01, epochs=50):
        X = np.array(inputs, dtype=float)
        y = np.array(outputs)

        self.weights = np.random.uniform(-0.01, 0.01, size=X.shape[1])
        self.bias = 0.0

        for epoch in range(epochs):
            for i in range(len(X)):
                x_i = X[i]
                pred = self.activation(self.weighting(x_i))
                error = y[i] - pred
                self.weights += learning_rate * error * x_i
                self.bias += learning_rate * error


           
```
