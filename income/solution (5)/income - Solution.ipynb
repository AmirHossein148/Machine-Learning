   اگر از کتابخانه‌ی <code>H2O</code> استفاده کنیم نیازی نیست تا خودمان به‌صورت دستی مقادیر گم‌شده را مدیریت کنیم و درخت تصمیم این کتابخانه مستقیماً از این نوع مقادیر پشتیبانی می‌کند.
    <br>
    اما اگر از <code>scikit-learn</code> استفاده کنیم جهت مدیریت مقادیر گم‌شده توجه به اینکه ویژگی‌هایی که شامل مقادیر گم‌شده هستند از نوع دسته‌ای هستند یکی از ساده‌ترین روش‌ها این است که آن‌ها را با پرتکرارترین مقدار (دسته) پُر کنیم. در کد زیر ابتدا پرتکرارترین دسته‌ی ویژگی را پیدا کرده سپس تمام مقادیر <code>NaN</code> آن ویژگی را هم در مجموعه‌ی آموزشی و هم آزمون را با این دسته جایگزین می‌کنیم. توجه شود که برای پیدا کردن پرتکرارترین دسته تنها از مجموعه‌ی آموزشی استفاده کرده‌ایم و آن را به مجموعه‌ی آزمون تعمیم داده‌ایم. از آن جهت که در این راه‌حل هر دو کتابخانه را امتحان خواهیم کرد، تغییرات مربوط به <code>scikit-learn</code> را در دو دیتافریم کپی‌شده اعمال می‌کنیم.
    </i>
</font>
</p>

```python
sk_train = train_data.copy()
sk_test = test_data.copy()
```

```python
cols = ['workclass', 'occupation', 'native.country']

for col in cols:
    most_frequent = sk_train[col].value_counts().sort_values(ascending=False).index[0]
    sk_train[col].fillna(most_frequent, inplace=True)
    sk_test[col].fillna(most_frequent, inplace=True)
    print('All the missing values in column', col, 'are replaced with', most_frequent)
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
    <i>
    حال به سراغ مدیریت ویژگی‌های دسته‌ای می‌رویم.
    باز هم اگر از کتابخانه‌ی <code>H2O</code> استفاده کنیم نیازی به مدیریت دستی ویژگی‌های دسته‌ای نداریم و این کتابخانه مستقیماً از ویژگی‌های دسته‌ای پشتیبانی می‌کند.
اما از آنجا که کتابخانه‌ی <code>scikit-learn</code> از مقادیر دسته‌ای پشتیبانی نمی‌کند نیاز است تا آن‌ها را به کمک روش‌هایی که در فصل مهندسی ویژگی آموختیم به مقادیر عددی تبدیل کنیم. در ابتدا ستون <code>income</code> را که همان متغیر هدف است به کمک <code>LabelEncoder</code> به برچسب‌های عددی تبدیل می‌کنیم.
    </i>
</font>
</p>

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
sk_train['income'] = le.fit_transform(sk_train['income'])
to_label = dict(zip(le.classes_, le.transform(le.classes_)))
to_class = dict(zip(le.transform(le.classes_), le.classes_))
print(to_label)
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
    <i>
    حال باید درباره سایر ویژگی‌های دسته‌ای تصمیم بگیریم. ساده‌ترین راه این است که به هر دسته یک عدد اختصاص دهیم و دسته‌ها را با این اعداد جایگزین کنیم. هرچند که این روش معایب خود را دارد اما به دلیل آن‌که قصد نداریم از روشی مانند One-Hot استفاده کرده و تعداد ویژگی‌ها را به شدت افزایش دهیم، به کمک تابع <code>OrdinalEncoder</code> از همین تکنیک استفاده خواهیم کرد. توجه داشته باشید که در نمونه‌های آزمون ممکن است دسته‌ای وجود داشته باشد که در هنگام آموزش آن را ندیده باشید. بنابراین می‌توانیم آرگومان <code>handle_unknown</code> را برابر <code>'use_encoded_value'</code> قرار دهید تا به جای این دسته‌ها مقداری که برای <code>unknown_value</code> تنظیم می‌کنید در نظر گرفته شود. برای این آرگومان می‌توانیم مقدار <code dir=ltr>-1</code> را در نظر بگیریم.
    </i>
</font>
</p>

```python
from sklearn.preprocessing import OrdinalEncoder

categorical_cols = sk_train.select_dtypes(include = "object").columns

enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

sk_train[categorical_cols] = enc.fit_transform(sk_train[categorical_cols])
sk_test[categorical_cols] = enc.transform(sk_test[categorical_cols])
```

```python
import h2o
h2o.init()
```

```python
hf_train = h2o.H2OFrame(train_data)
hf_test = h2o.H2OFrame(test_data)
```

<font face="vazirmatn" size=3>
    <i>
    سپس باید یک لیست از ویژگی‌ها بسازیم. برای این کار نام ستون‌های فریم‌ را گرفته و ستون هدف را از آن حذف می‌کنیم.
    </i>

```python
TARGET = 'income'
features = list(hf_train.columns)
features.remove(TARGET)
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazirmatn" size=3>
    <i>
    اکنون می‌توانیم پارامترهای موردنظر را مقداردهی کرده و مدل را آموزش دهیم. در اینجا با مقداردهی <code>True</code> به آرگومان <code>balance_classes</code> می‌توانیم از روش نمونه‌افزایی به کمک تکرار نمونه‌های اقلیت استفاده کنیم. البته این کتابخانه آرگومان‌های دیگری را جهت مدیریت توازن داده‌ها در اختیار شما قرار می‌دهد اما جهت حفظ سادگی از استفاده‌ی آن‌ها پرهیز کرده‌ایم.
    </i>
</font>

```python
from h2o.estimators import H2ORandomForestEstimator

model = H2ORandomForestEstimator(ntrees=1, sample_rate=1, mtries=-2, max_depth=10,
                                 min_rows=8, min_split_improvement=1e-4,
                                 balance_classes=True, nfolds=5)

model.train(x=features, y=TARGET, training_frame=hf_train)
```
